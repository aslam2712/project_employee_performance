```python
import pandas as pd
import numpy as np
```

```python
from random import seed
from random import randrange
```

```python
from csv import reader
from math import sqrt
```

```python
seed (1)
filename = 'INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xls'
```

```python
dataset = pd.read_excel(filename)
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = dataset.iloc[:,2:26]
X = pd.get_dummies(X)
```

```python
X.head()
```

```python
y = dataset.iloc[:,-1]
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.33, random_state=25)
```

```python
print (X_train.shape)
print (y_train.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators= 1000, random_state=100)

# Train the model on training data
```

```python
rf.fit(X_train,y_train)
```

```python
y_test = y_test.reshape(-1,1)
```

```python
predictions =rf.predict(X_test)
errors = abs(predictions - y_test)
```

```python
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / y_test)
```

```python
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
